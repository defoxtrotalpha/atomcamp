```python
dataset_id = "samsum"
from datasets import load_dataset

dataset = load_dataset(dataset_id)
print(f"Train dataset size: {len(dataset['train'])}")
print(f"Test dataset size: {len(dataset['test'])}")
```

```python
from random import randrange

#Selects a random sample from the training dataset.
sample = dataset['train'][randrange(len(dataset["train"]))]
print(f"dialogue: \n{sample['dialogue']}\n---------------")
print(f"summary: \n{sample['summary']}\n---------------")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_id = "google/flan-t5-base"
#Loads the tokenizer for the specified model.
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
from datasets import concatenate_datasets
#Concatenates the training and test datasets.
concat_data =  concatenate_datasets([dataset["train"], dataset["test"]])

#Tokenizes the dialogues, truncating them to fit the model's input requirements.
tokenized_inputs = concat_data.map(lambda x: tokenizer(x["dialogue"], truncation=True), batched=True, remove_columns=["dialogue", "summary"])
max_source_length = max([len(x) for x in tokenized_inputs["input_ids"]])
print(f"Max source length: {max_source_length}")


tokenized_targets = concat_data.map(lambda x: tokenizer(x["summary"], truncation=True), batched=True, remove_columns=["dialogue", "summary"])
max_target_length = max([len(x) for x in tokenized_targets["input_ids"]])
print(f"Max target length: {max_target_length}")
```

```python
def preprocess_function(sample,padding="max_length"):
  # Adds a prefix "summarize: " to each dialogue.
  inputs = ["summarize: " + item for item in sample["dialogue"]]

  # Tokenizes the inputs, padding or truncating them to the maximum source length.
  model_inputs = tokenizer(inputs, max_length=max_source_length, padding=padding, truncation=True)
  labels = tokenizer(text_target=sample["summary"], max_length=max_target_length, padding=padding, truncation=True)

  # Replaces padding token IDs with -100 in the labels to ignore them during loss computation.
  if padding == "max_length":
        labels["input_ids"] = [
            [(l if l != tokenizer.pad_token_id else -100) for l in label] for label in labels["input_ids"]
        ]
  # Adds the processed labels to the model inputs.
  model_inputs["labels"] = labels["input_ids"]
  return model_inputs

# Applies the preprocessing function to the dataset, removing unnecessary columns.
tokenized_dataset = dataset.map(preprocess_function, batched=True, remove_columns=["dialogue", "summary", "id"])

# Prints the keys of the tokenized dataset.
print(f"Keys of tokenized dataset: {list(tokenized_dataset['train'].features)}")
```

```python
from transformers import AutoModelForSeq2SeqLM

model_id = "google/flan-t5-base"

model = AutoModelForSeq2SeqLM.from_pretrained(model_id)
```

```python
import evaluate
import nltk
import numpy as np
from nltk.tokenize import sent_tokenize
nltk.download("punkt") # Downloads the "punkt" tokenizer models for sentence tokenization.

metric = evaluate.load("rouge")
#BLUE METEOR SPICE CiDER

def postprocess_text(preds, labels):
  preds = [pred.strip() for pred in preds]
  labels = [label.strip() for label in labels]

# Splits each prediction into sentences and joins them with newline characters.
  preds = ["\n".join(sent_tokenize(pred)) for pred in preds]
  labels = ["\n".join(sent_tokenize(label)) for label in labels]

  return preds, labels

def compute_metrics(eval_preds):
  preds, labels = eval_preds
  #  If preds is a tuple, extracts the first element.
  if isinstance(preds, tuple):
    preds = preds[0]

# Decodes the predictions using the tokenizer, skipping special tokens.
  decode_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)

# Replaces -100 in labels with the padding token ID.
  lables = np.where(labels != -100, labels, tokenizer.pad_token_id)

# Decodes the labels using the tokenizer, skipping special tokens.
  decode_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

# Post-processes the decoded predictions and labels.
  decoded_preds, decoded_lables = postprocess_text(decoded_preds, decoded_lables)

# Computes the ROUGE metrics using the decoded predictions and labels.
  result = metric.compute(predictions = decoded_preds, references=decoded_lables, use_stemmer=True)

#  Calculates the length of each prediction.
  prediction_lens = [np.count_nonzero(pred != tokenizer.pad_token_id) for pred in preds]

# Adds the average prediction length to the result.
  result["gen_len"] = np.mean(prediction_lens)
  return result
```

```python
from transformers import DataCollatorForSeq2Seq

# Creates a data collator for sequence-to-sequence models, ensuring padding is a multiple of 8.
label_pad_token_id = -100
data_collator = DataCollatorForSeq2Seq(
    tokenizer,
    model=model,
    label_pad_token_id=label_pad_token_id,
    pad_to_multiple_of=8
)
```

```python
from huggingface_hub import HfFolder
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments

repository_id = f"{model_id.split('/')[1]}-{dataset_id}"
training_args=Seq2SeqTrainingArguments(
    output_dir=repository_id,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    predict_with_generate=True,
    fp16=False, # Overflows with fp16
    learning_rate=5e-5,
    num_train_epochs=1,
    # logging & evaluation strategies
    logging_dir=f"{repository_id}/logs",
    logging_strategy="steps",
    logging_steps=500,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=2,
    load_best_model_at_end=True,
    # metric_for_best_model="overall_f1",
    # push to hub parameters
    report_to="tensorboard",
    push_to_hub=False,
    hub_strategy="every_save",
    hub_model_id=repository_id,
    hub_token=HfFolder.get_token(),
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset = tokenized_dataset["train"],
    eval_dataset = tokenized_dataset["test"],
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
# Saves the tokenizer to the specified repository ID.
tokenizer.save_pretrained(repository_id)
trainer.create_model_card()
# Pushes the trained model and tokenizer to the Hugging Face Hub.
trainer.push_to_hub()
```
