#Discretization#
quantile based binning. Making categories out of a continuous data

```python
cats= ['Intern', 'Junior', 'Mid-level', 'Senior']
df['AgeRange']=pd.qcut(df['Age'], q=4, labels=cats)
df[['AgeRange', 'Age']]
```

```python
df['AgeRange'].value_counts()
```

```python
df.columns
print(df['MonthlyIncome'].max())
print(df['MonthlyIncome'].min())
```

cut is used for custom bins while qcut is used for quantile based bins

```python
cut_labels= ['copper', 'silver', 'gold', 'platinum', 'diamond']
bins = [1000,4000,6000,12000,15000,20000]
df['Income Package']=pd.cut(df['MonthlyIncome'], bins=bins, labels=cut_labels)
df[['Income Package','MonthlyIncome']]
```

```python
plt.hist(df['MonthlyIncome'])
plt.show() #This plot is a little tough to intrepret
```

```python
plt.hist(df['Income Package'])
plt.show() #This plot is a easier to intrepret
```
