### Key Insights  
1. Data appraently has no null values but we do see '--' values in many columns
2. Numerical columns also have object datatype because they have '--' as null value  

We need to treat these numerical columns so that we can get more detail about the data. 

```python
df1=df.replace('..',pd.NA)
df1.tail()
```

Now we have to convert all the columns that should be numerical but are object into float

```python
numeric_columns = df1.loc[:, '1990':'2018'].apply(pd.to_numeric, errors='coerce')
numeric_columns.info()
```

Concatinating object columns with numeric columns

```python
# Concatenating DataFrames horizontally (along columns)
df2 = pd.concat([df1[['greenhouse gas', 'sector']], numeric_columns], axis=1)
df2.info()
```

Now we have to remove rows with all null entries

```python
df_cleaned = df2.dropna(thresh=len(df.columns) - 2)

from IPython.display import display
with pd.option_context('display.max_rows', 160):
    display(df_cleaned)
```

## Let's check the overall emissions in each decade

```python
decades = {
    '1990s': 0,
    '2000s': 0,
    '2010s': 0
}

for year in df_cleaned.loc[:, '1990': '2018']:  # Iteratig over the years
    if year.startswith('199'):
        decades['1990s'] += df_cleaned[year].sum()
    elif year.startswith('200'):
        decades['2000s'] += df_cleaned[year].sum()
    elif year.startswith('201'):
        decades['2010s'] += df_cleaned[year].sum()

# Creating DataFrame from the dictionary
decade_totals = pd.DataFrame(decades, index=[0])  #For single row

print(decade_totals)
```

```python
import matplotlib.pyplot as plt

# Plotting
decade_totals.plot(kind='bar', legend=None)
plt.title('Total Emissions by Decade')
plt.xlabel('Decade')
plt.ylabel('Total Emissions')
plt.xticks(rotation=0)  # Rotate x-labels for better readability
plt.show()
```

## Binning
We will consider emissions in year 2018 for this purpose

```python
cats= ['low', 'medium', 'high', 'very high']
df_summary.loc[:,'EmissionsRange']=pd.qcut(df_summary['2018'], q=4, labels=cats)
df_summary[['sector','EmissionsRange', '2018']]
```

1. Here we can see the range of sectors with low and very high emissions.
2. Data is categorized using quantile based binning (Median treated as mid)

We will try to the same using custom bins now

```python
df_summary['2018'].max()
```

```python
df_summary['2018'].min() 
```

### The process of removing CO2 from the atmosphere is known as negative emissions, because more CO2 is being taken out of the atmosphere than added into it.

```python
bins = [-42000,-10000,15000,20000,64000];
cats= ['low', 'medium', 'high', 'very high']
df_summary.loc[:,'EmissionsRange']=pd.cut(df_summary['2018'], bins=bins, labels=cats)
df_summary[['sector','EmissionsRange', '2018']]
```

Here we can see that the result is a bit different because we have used custom discretization
