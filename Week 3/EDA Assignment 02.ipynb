Q4. Determine the missing values in the whole dataset and analyze missing values in each
column.

```python
missing_count = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_analysis = pd.DataFrame({'Missing Value Count': missing_count, 'Missing Value Percentage': missing_percentage})

print("\nMissing Values Analysis:\n", missing_analysis)
```

5. How many clinical trials started after January 1, 2020? 

```python
df['Start Date'] #Inspecting column
```

Data is not sorted and it is of object dtype. First we will convert it into datetime dtype

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], format='mixed', errors='coerce')
df['Start Date']
```

```python
df_temp = df[df['Start Date'] > pd.to_datetime('2020-01-01')]
df_temp
```
